### Этап 5: Обучение моделей и подбор гиперпараметров
Импорт и обучение моделей

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Список категориальных признаков
categorical_features = ['State', 'Response', 'Coverage', 'Education', 'EmploymentStatus', 'Gender', 'Location Code', 'Marital Status', 'Policy Type', 'Policy', 'Renew Offer Type', 'Sales Channel', 'Vehicle Class', 'Vehicle Size']

# Преобразование категориальных признаков в числовые метки
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Разделение данных на признаки и целевую переменную
X = data.drop(columns=['Customer Lifetime Value', 'Customer', 'Effective To Date', 'LTV_binary'])
y = data['LTV_binary']

# Стандартизация числовых признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

* Категориальные признаки были преобразованы в числовые с помощью LabelEncoder.
* Стандартизированы числовые признаки с помощью StandardScaler.

### 6.Построение и оценка моделей:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate

# Создание модели логистической регрессии
log_reg = LogisticRegression(C=0.8, random_state=42, max_iter=500)

# Проведение кросс-валидации
cv_log_reg = cross_validate(log_reg, X_scaled, y, cv=5, scoring='roc_auc')

# Вывод результатов кросс-валидации
print("Результаты Логистической регрессии:", cv_log_reg['test_score'].mean())

```

```python
# Решающее дерево
from sklearn.tree import DecisionTreeClassifier


dec_tree = DecisionTreeClassifier(random_state = 42) # Можно найти более оптимальные гиперпараметры

cv_dec_tree = cross_validate(dec_tree, X_sc, y, cv = 5, scoring = 'roc_auc')

print("Результаты решающего дерева:", sum(cv_dec_tree['test_score']) / len(cv_dec_tree['test_score']))
```

```python
# Случайный лес
from sklearn.ensemble import RandomForestClassifier


ran_for = RandomForestClassifier(random_state = 42) # Можно найти более оптимальные гиперпараметры

cv_ran_for = cross_validate(ran_for, X_sc, y, cv = 5, scoring = 'roc_auc')

print("Результаты решающего дерева:", sum(cv_ran_for['test_score']) / len(cv_ran_for['test_score']))
```

* Использованы три модели: логистическая регрессия, решающее дерево и случайный лес.
* Проведена кросс-валидация с оценкой по метрике ROC-AUC.

### 7. Анализ важности признаков:

```python
# Обучение модели случайного леса на всех данных
ran_for.fit(X_scaled, y)

# Важность признаков
importances = ran_for.feature_importances_
feature_names = X.columns
feature_importance = pd.Series(importances, index=feature_names).sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=feature_importance, y=feature_importance.index)
plt.title('Важность признаков для случайного леса')
plt.show()
```

* На основе обученной модели случайного леса была проведена оценка важности признаков.

Оценка моделей
Логистическая регрессия
* Плюсы:

    * Простота и интерпретируемость.
    * Быстрота обучения и предсказаний.
    * Хорошо работает с линейно разделимыми данными.
* Минусы:

    * Ограниченная способность захватывать нелинейные зависимости.
    * Сравнительно низкий показатель ROC-AUC (0.7497), что свидетельствует о недостаточной эффективности модели для данной задачи.
Решающее дерево
* Плюсы:
    * Легкость интерпретации благодаря визуализации дерева решений.
    * Возможность захвата нелинейных зависимостей.
    * Хороший показатель ROC-AUC (0.9877).
* Минусы:
    * Склонность к переобучению, особенно на малых выборках или при отсутствии регуляризации.
    *Модель может быть нестабильной, так как небольшие изменения в данных могут сильно изменить структуру дерева.
Случайный лес
* Плюсы:
    * Высокая точность и устойчивость к переобучению благодаря ансамблевому подходу.
    * Способность обрабатывать большие объемы данных и учитывать сложные зависимости.
    * Лучший показатель ROC-AUC (0.9992), что свидетельствует о высокой эффективности модели для данной задачи.
* Минусы:
    * Меньшая интерпретируемость по сравнению с логистической регрессией и решающим деревом.
    * Более длительное время обучения и предсказаний по сравнению с простыми моделями.

3. Обработка сложных зависимостей: Случайный лес эффективно обрабатывает нелинейные зависимости и взаимодействия между признаками, что делает его подходящим для данной задачи с большим количеством категориальных и числовых признаков.
